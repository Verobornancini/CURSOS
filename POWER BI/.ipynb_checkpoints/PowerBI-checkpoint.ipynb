### **Buenas pr치cticas para modelado de datos**
* Un modelo contar치 con tablas conectadas por medio de relaciones
* El modelado de datos consiste en: Cargar las tablas, relacionarlas, transformar los datos con expresiones DAX
* Cada entidad de negocio debe tener su propia tabla, ejemplo: Si analizamos escuelas podriamos tener tablas como; Calificaciones, Asistencias, Maestros, Alumnos, Materias, Ex치menes.
* Las entidades de negocio se diferencian en dos tipos:

**FUNCIONES DAX**
* **CALCULATE:** Esta funci칩n permite realizar operaciones teniendo en cuenta filtros o condiciones, la funci칩n CALCULATE se usa para filtrar mediante un argumento que pasamos a la funci칩n. Esto es lo que se conoce como funciones anidadas. La funci칩n CALCULATE tiene al menos dos argumentos. El primero es la expresi칩n que se eval칰a; el segundo, un filtro, por ejemplo:
    * Efectivo Inicial = CALCULATE(SUM('Flujo de Caja'[Value]), FILTER('Flujo de Caja', 'Flujo de Caja'[Meses]= "Dinero liquido al inicio (caja y bancos)"))
    * Admite operaciones sobre alguna columna (SUM(columna1)).
 
* ALL: Se utiliza como un filtro dentro de CALCULATE, se usa para incluir a todos los registros de una tabla, por ejemplo:
    * All Total Ordenes = CALCULATE([Total Ordenes], ALL(FactSales))

* **FILLTER:** Primero aplica el filtro que especifiquemos en la tabla,y luego hace el c치lculo que definimos en el primer t칠rmino antes de la coma. En el filtro, se coloca el nombre de la tabla y luego de una coma el campo con la expresion que quiero filtrar.
    * Total Ordenes devueltas Multiples Filter = CALCULATE([Total Ordenes],FILTER(FactSales, FactSales[ReturnQuantity]>0))
    * Fillter adem치s nos permite usar como filtros a m칠tricas previamente creadas por nosotros, que no podemos usarlas con calculate si no ponemos a filter en el filtro propiamente dicho.
 
* **RELATED:** Permite tomar campos de otras tablas, por ejemplo:
    * Precio unitario: RELATED(Dimproduct [UnitPrice]) -> Traigo a una nueva tabla la columna de unitprice y la renombro.
        
* **SUMX:** Para hacer sumatorias de multiplicaciones de dos campos de una tabla o de distintas tablas. Por ejemplo:
    * Total Ingresos mismatabla = SUMX(factsales, factsales[cantidad total] * factsales[precio unitario])
    * Total Ingresos dostablas = SUMX(FactSales, FactSales[Cantidad Total] * RELATED(DimProduct[UnitPrice]))
    
* **RANKX:** Para crer un ranking de medidas. Por ejemplo:
        * Rank_ingresos_tiendas= RANKX(ALL(Dimstore), [Total ingresos])
    * Si tenemos registros en blancos podemos filtrarlos para que no se aplique en ellos la formula de rank.
